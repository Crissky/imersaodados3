## Lendo CSV zipado e exibindo dados

## Renomeando Coluna

Quando o argumento "inplace=True" não é usado, a alteração do nome da Coluna não é feita no DataFrame, mas em uma cópia que a função "rename" retorna.

```python
mapa = {'droga': 'composto'}
df.rename(columns=mapa)
```

```python
df.head()
```

```python
df.rename(columns=mapa, inplace=True)
```

```python
df.head()
```

## Gráficos com o Seaborn
### Importando Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set() # Carrega as configurações padrão do seaborn (tema).
# plt.figure(figsize=(10,8)) # Define o tamanho dos gráficos.
plt.rcParams["figure.figsize"] = (10,8) # Define o tamanho dos gráficos.
```

## Selecionando os 5 elementos que mais se repetem da Coluna composto

```python
compostos = df.composto.value_counts()[:5]

compostos
```

## Pegando somente os índices (labels) dos 5 elementos que mais se repetem da Coluna composto

```python
cod_compostos = compostos.index

cod_compostos
```

## Query que filtra o Dataframe com os 5 elementos que mais se repetem da Coluna composto.

```python
query_composto = df.query('composto in @cod_compostos')

query_composto
```

## Exibindo Gráfico com todos os compostos

```python
sns.countplot(x='composto', data=df).set_title('Top 5 Compostos')
```

## Exibindo gráfico com os 5 elementos que mais se repetem da Coluna composto

O gráfico pode ser armazenado em uma variável e exibido posteriormente com o "plt.show()".

```python
ax = sns.countplot(x='composto', data=query_composto)
ax.set_title('Top 5 Compostos')
plt.show()
```
